```python
# ============================================================================
# STEP 1: DEFINE CORE KPIs
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set styling
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Load the data
df = pd.read_csv('sales_data.csv')

# Data Cleaning
print("=" * 60)
print("DATA CLEANING & PREPARATION")
print("=" * 60)

# Handle missing values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce').fillna(0)
df['Return'] = pd.to_numeric(df['Return'], errors='coerce').fillna(0)
df['SatisfactionScore'] = pd.to_numeric(df['SatisfactionScore'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')

# Fill missing Region with 'Unknown'
df['Region'] = df['Region'].fillna('Unknown')

# Create Date column from the time-based Date field (simulate date for analysis)
df['TransactionID'] = df['CustomerID'].astype(str) + '_' + df['ProductID'].astype(str)

print(f"Total Records: {len(df)}")
print(f"Missing Values:\n{df.isnull().sum()}")
print("\n" + "=" * 60)

# ============================================================================
# CALCULATE CORE KPIs
# ============================================================================

def calculate_kpis(dataframe):
    """Calculate all core KPIs with detailed breakdowns"""
    
    kpis = {}
    
    # KPI 1: Average Order Value (AOV)
    kpis['AOV'] = dataframe['Sales'].mean()
    
    # KPI 2: Return Rate
    total_orders = len(dataframe)
    total_returns = dataframe['Return'].sum()
    kpis['Return_Rate'] = (total_returns / total_orders) * 100
    
    # KPI 3: Customer Satisfaction Score
    kpis['Avg_Satisfaction'] = dataframe['SatisfactionScore'].mean()
    
    # KPI 4: Discount Impact Ratio
    discounted_sales = dataframe[dataframe['Discount'] > 0]['Sales'].sum()
    total_sales = dataframe['Sales'].sum()
    kpis['Discount_Impact_Ratio'] = (discounted_sales / total_sales) * 100 if total_sales > 0 else 0
    
    # KPI 5: VIP Customer Revenue Share
    vip_sales = dataframe[dataframe['CustomerType'] == 'VIP']['Sales'].sum()
    kpis['VIP_Revenue_Share'] = (vip_sales / total_sales) * 100 if total_sales > 0 else 0
    
    # Additional Metrics
    kpis['Total_Revenue'] = total_sales
    kpis['Total_Transactions'] = total_orders
    kpis['Total_Returns'] = total_returns
    
    return kpis

# Calculate KPIs
kpis = calculate_kpis(df)

# Display KPIs
print("\n" + "=" * 60)
print("CORE KPIs DASHBOARD")
print("=" * 60)

for kpi, value in kpis.items():
    if 'Rate' in kpi or 'Share' in kpi or 'Ratio' in kpi:
        print(f"{kpi}: {value:.2f}%")
    elif 'Revenue' in kpi or 'AOV' in kpi:
        print(f"{kpi}: ${value:,.2f}")
    else:
        print(f"{kpi}: {value:.2f}")

print("=" * 60)
```

```python
# ============================================================================
# STEP 2: DEEP-DIVE ANALYSIS - CUSTOMER SEGMENTATION
# ============================================================================

print("\n" + "=" * 60)
print("DEEP-DIVE ANALYSIS: CUSTOMER SEGMENTATION")
print("=" * 60)

# ----------------------------------------------------------------------------
# 2.1 Segment Performance Analysis
# ----------------------------------------------------------------------------

def analyze_customer_segments(dataframe):
    """Analyze performance by customer type segments"""
    
    segment_analysis = dataframe.groupby('CustomerType').agg({
        'Sales': ['sum', 'mean', 'count'],
        'Return': 'sum',
        'SatisfactionScore': 'mean',
        'Discount': 'mean',
        'Quantity': 'sum'
    }).round(2)
    
    segment_analysis.columns = ['Total_Sales', 'Avg_Order_Value', 'Transaction_Count', 
                                 'Total_Returns', 'Avg_Satisfaction', 'Avg_Discount', 
                                 'Total_Quantity']
    
    # Calculate Return Rate per segment
    segment_analysis['Return_Rate'] = (segment_analysis['Total_Returns'] / 
                                        segment_analysis['Transaction_Count']) * 100
    
    # Calculate Revenue Share
    total_revenue = segment_analysis['Total_Sales'].sum()
    segment_analysis['Revenue_Share'] = (segment_analysis['Total_Sales'] / 
                                          total_revenue) * 100
    
    return segment_analysis

segment_perf = analyze_customer_segments(df)
print("\n--- Customer Segment Performance ---")
print(segment_perf)

# ----------------------------------------------------------------------------
# 2.2 Regional Analysis by Segment
# ----------------------------------------------------------------------------

def regional_segment_analysis(dataframe):
    """Analyze regional performance by customer segment"""
    
    regional_seg = dataframe.groupby(['Region', 'CustomerType']).agg({
        'Sales': 'sum',
        'TransactionID': 'count',
        'SatisfactionScore': 'mean'
    }).reset_index()
    
    regional_seg.columns = ['Region', 'CustomerType', 'Total_Sales', 
                            'Transactions', 'Avg_Satisfaction']
    
    return regional_seg

regional_seg = regional_segment_analysis(df)

# ----------------------------------------------------------------------------
# 2.3 Product Category Performance by Segment
# ----------------------------------------------------------------------------

def category_segment_analysis(dataframe):
    """Analyze product category performance by customer segment"""
    
    cat_seg = dataframe.groupby(['ProductCategory', 'CustomerType']).agg({
        'Sales': 'sum',
        'Return': 'sum',
        'Quantity': 'sum'
    }).reset_index()
    
    cat_seg['Return_Rate'] = (cat_seg['Return'] / 
                              cat_seg.groupby(['ProductCategory', 'CustomerType'])
                              ['Return'].transform('count')) * 100
    
    return cat_seg

cat_seg = category_segment_analysis(df)

# ----------------------------------------------------------------------------
# 2.4 Sales Channel Analysis
# ----------------------------------------------------------------------------

def channel_analysis(dataframe):
    """Analyze sales channel performance"""
    
    channel_perf = dataframe.groupby('SalesChannel').agg({
        'Sales': ['sum', 'mean'],
        'TransactionID': 'count',
        'SatisfactionScore': 'mean',
        'Return': 'sum'
    }).round(2)
    
    channel_perf.columns = ['Total_Sales', 'Avg_Order_Value', 'Transactions', 
                            'Avg_Satisfaction', 'Total_Returns']
    channel_perf['Return_Rate'] = (channel_perf['Total_Returns'] / 
                                   channel_perf['Transactions']) * 100
    
    return channel_perf

channel_perf = channel_analysis(df)
print("\n--- Sales Channel Performance ---")
print(channel_perf)

# ============================================================================
# VISUALIZATIONS FOR DEEP-DIVE ANALYSIS
# ============================================================================

# Create comprehensive visualization dashboard
fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=('Revenue by Customer Segment', 'Return Rate by Segment',
                    'Sales by Region & Segment', 'Satisfaction Score Distribution',
                    'Sales Channel Performance', 'Top Product Categories'),
    specs=[[{"type": "bar"}, {"type": "bar"}],
           [{"type": "heatmap"}, {"type": "histogram"}],
           [{"type": "pie"}, {"type": "bar"}]]
)

# 1. Revenue by Customer Segment
segment_revenue = df.groupby('CustomerType')['Sales'].sum().reset_index()
fig.add_trace(
    go.Bar(x=segment_revenue['CustomerType'], y=segment_revenue['Sales'],
           name='Revenue', marker_color=['#FF6B6B', '#4ECDC4', '#45B7D1']),
    row=1, col=1
)

# 2. Return Rate by Segment
return_by_seg = df.groupby('CustomerType').apply(
    lambda x: (x['Return'].sum() / len(x)) * 100
).reset_index()
return_by_seg.columns = ['CustomerType', 'Return_Rate']
fig.add_trace(
    go.Bar(x=return_by_seg['CustomerType'], y=return_by_seg['Return_Rate'],
           name='Return Rate', marker_color=['#FF8E72', '#6BCB77', '#4D96FF']),
    row=1, col=2
)

# 3. Regional Heatmap
region_seg_pivot = df.pivot_table(values='Sales', index='Region', 
                                   columns='CustomerType', aggfunc='sum', fill_value=0)
fig.add_trace(
    go.Heatmap(z=region_seg_pivot.values, x=region_seg_pivot.columns, 
               y=region_seg_pivot.index, colorscale='Blues'),
    row=2, col=1
)

# 4. Satisfaction Score Distribution
fig.add_trace(
    go.Histogram(x=df['SatisfactionScore'], nbinsx=10, name='Satisfaction',
                 marker_color='#FFD93D'),
    row=2, col=2
)

# 5. Sales Channel Performance
channel_sales = df.groupby('SalesChannel')['Sales'].sum().reset_index()
fig.add_trace(
    go.Pie(labels=channel_sales['SalesChannel'], values=channel_sales['Sales'],
           name='Channel Share'),
    row=3, col=1
)

# 6. Top Product Categories
cat_sales = df.groupby('ProductCategory')['Sales'].sum().nlargest(5).reset_index()
fig.add_trace(
    go.Bar(x=cat_sales['ProductCategory'], y=cat_sales['Sales'],
           name='Top Categories', marker_color='#6C5CE7'),
    row=3, col=2
)

fig.update_layout(height=1200, width=1400, 
                  title_text='📊 Customer Segmentation Deep-Dive Analysis',
                  showlegend=False)
fig.show()

# Save the figure
fig.write_html('customer_segmentation_analysis.html')
print("\nDeep-dive analysis visualization saved as 'customer_segmentation_analysis.html'")
```

```python
# ============================================================================
# STEP 3: INTERACTIVE DASHBOARD BUILD (FIXED VERSION)
# ============================================================================
# Save this code as: dashboard.py
# Run with: streamlit run dashboard.py

dashboard_code = '''
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set UTF-8 encoding for Windows compatibility
import sys
import io
sys.stdout = io.TextIOWrapper(sys.stdout.buffer, encoding='utf-8')

# Page Configuration
st.set_page_config(
    page_title="Sales Analytics Dashboard",
    page_icon=":bar_chart:",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS
st.markdown("""
    <style>
    .main-header {
        font-size: 2.5rem;
        font-weight: bold;
        color: #1E88E5;
        text-align: center;
        padding: 1rem 0;
    }
    .kpi-card {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        border-radius: 10px;
        padding: 1.5rem;
        color: white;
        text-align: center;
    }
    .metric-value {
        font-size: 2rem;
        font-weight: bold;
    }
    .metric-label {
        font-size: 0.9rem;
        opacity: 0.9;
    }
    </style>
""", unsafe_allow_html=True)

# Load Data
@st.cache_data
def load_data():
    df = pd.read_csv('sales_data.csv')
    # Data Cleaning
    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
    df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce').fillna(0)
    df['Return'] = pd.to_numeric(df['Return'], errors='coerce').fillna(0)
    df['SatisfactionScore'] = pd.to_numeric(df['SatisfactionScore'], errors='coerce')
    df['Region'] = df['Region'].fillna('Unknown')
    return df

df = load_data()

# Header
st.markdown('<p class="main-header">Sales Analytics Dashboard</p>', 
            unsafe_allow_html=True)
st.markdown("---")

# Sidebar Filters
st.sidebar.header("Filters")

# Region Filter
regions = ['All'] + list(df['Region'].unique())
selected_region = st.sidebar.selectbox("Select Region", regions)

# Customer Type Filter
customer_types = ['All'] + list(df['CustomerType'].unique())
selected_customer = st.sidebar.selectbox("Customer Type", customer_types)

# Sales Channel Filter
channels = ['All'] + list(df['SalesChannel'].unique())
selected_channel = st.sidebar.selectbox("Sales Channel", channels)

# Product Category Filter
categories = ['All'] + list(df['ProductCategory'].unique())
selected_category = st.sidebar.selectbox("Product Category", categories)

# Apply Filters
filtered_df = df.copy()
if selected_region != 'All':
    filtered_df = filtered_df[filtered_df['Region'] == selected_region]
if selected_customer != 'All':
    filtered_df = filtered_df[filtered_df['CustomerType'] == selected_customer]
if selected_channel != 'All':
    filtered_df = filtered_df[filtered_df['SalesChannel'] == selected_channel]
if selected_category != 'All':
    filtered_df = filtered_df[filtered_df['ProductCategory'] == selected_category]

# Calculate KPIs
total_revenue = filtered_df['Sales'].sum()
total_transactions = len(filtered_df)
aov = filtered_df['Sales'].mean()
return_rate = (filtered_df['Return'].sum() / total_transactions) * 100 if total_transactions > 0 else 0
avg_satisfaction = filtered_df['SatisfactionScore'].mean()
vip_sales = filtered_df[filtered_df['CustomerType']=='VIP']['Sales'].sum()
vip_share = (vip_sales / total_revenue * 100) if total_revenue > 0 else 0

# KPI Cards
st.subheader("Core KPIs")
kpi1, kpi2, kpi3, kpi4, kpi5 = st.columns(5)

with kpi1:
    st.markdown(f"""
    <div class="kpi-card">
        <div class="metric-value">${total_revenue:,.0f}</div>
        <div class="metric-label">Total Revenue</div>
    </div>
    """, unsafe_allow_html=True)

with kpi2:
    st.markdown(f"""
    <div class="kpi-card">
        <div class="metric-value">${aov:,.0f}</div>
        <div class="metric-label">Avg Order Value</div>
    </div>
    """, unsafe_allow_html=True)

with kpi3:
    st.markdown(f"""
    <div class="kpi-card">
        <div class="metric-value">{return_rate:.1f}%</div>
        <div class="metric-label">Return Rate</div>
    </div>
    """, unsafe_allow_html=True)

with kpi4:
    st.markdown(f"""
    <div class="kpi-card">
        <div class="metric-value">{avg_satisfaction:.1f}</div>
        <div class="metric-label">Avg Satisfaction</div>
    </div>
    """, unsafe_allow_html=True)

with kpi5:
    st.markdown(f"""
    <div class="kpi-card">
        <div class="metric-value">{vip_share:.1f}%</div>
        <div class="metric-label">VIP Revenue Share</div>
    </div>
    """, unsafe_allow_html=True)

st.markdown("---")

# Main Charts
col1, col2 = st.columns(2)

with col1:
    st.subheader("Revenue by Customer Segment")
    seg_rev = filtered_df.groupby('CustomerType')['Sales'].sum().reset_index()
    fig1 = px.bar(seg_rev, x='CustomerType', y='Sales', 
                  color='CustomerType',
                  color_discrete_sequence=['#FF6B6B', '#4ECDC4', '#45B7D1'])
    fig1.update_layout(showlegend=False, height=400)
    st.plotly_chart(fig1, use_container_width=True)

with col2:
    st.subheader("Revenue by Region")
    reg_rev = filtered_df.groupby('Region')['Sales'].sum().reset_index()
    fig2 = px.pie(reg_rev, values='Sales', names='Region',
                  color_discrete_sequence=px.colors.qualitative.Set3)
    fig2.update_layout(height=400)
    st.plotly_chart(fig2, use_container_width=True)

# Second Row Charts
col3, col4 = st.columns(2)

with col3:
    st.subheader("Sales Channel Performance")
    chan_perf = filtered_df.groupby('SalesChannel').agg({
        'Sales': 'sum',
        'SatisfactionScore': 'mean'
    }).reset_index()
    fig3 = px.bar(chan_perf, x='SalesChannel', y='Sales',
                  color='SatisfactionScore',
                  color_continuous_scale='RdYlGn')
    fig3.update_layout(height=400)
    st.plotly_chart(fig3, use_container_width=True)

with col4:
    st.subheader("Top Product Categories")
    cat_rev = filtered_df.groupby('ProductCategory')['Sales'].sum().nlargest(5).reset_index()
    fig4 = px.bar(cat_rev, x='Sales', y='ProductCategory', orientation='h',
                  color='Sales', color_continuous_scale='Blues')
    fig4.update_layout(height=400)
    st.plotly_chart(fig4, use_container_width=True)

# Deep Dive Section
st.markdown("---")
st.subheader("Deep-Dive Analysis: Customer Segmentation")

# Segmentation Table
st.write("### Segment Performance Metrics")
seg_metrics = filtered_df.groupby('CustomerType').agg({
    'Sales': ['sum', 'mean'],
    'Return': 'sum',
    'SatisfactionScore': 'mean',
    'CustomerID': 'count'
}).round(2)
seg_metrics.columns = ['Total_Sales', 'Avg_Order_Value', 'Total_Returns', 
                       'Avg_Satisfaction', 'Transactions']
seg_metrics['Return_Rate'] = (seg_metrics['Total_Returns'] / 
                              seg_metrics['Transactions']) * 100
st.dataframe(seg_metrics.style.format({
    'Total_Sales': '${:,.2f}',
    'Avg_Order_Value': '${:,.2f}',
    'Return_Rate': '{:.2f}%'
}))

# Satisfaction Distribution
st.write("### Satisfaction Score Distribution")
fig5 = px.histogram(filtered_df, x='SatisfactionScore', nbins=10,
                    color='CustomerType', barmode='overlay',
                    opacity=0.7)
fig5.update_layout(height=400)
st.plotly_chart(fig5, use_container_width=True)

# Data Export
st.markdown("---")
st.subheader("Export Data")
if st.button("Download Filtered Data"):
    csv = filtered_df.to_csv(index=False)
    st.download_button(
        label="Download CSV",
        data=csv,
        file_name='filtered_sales_data.csv',
        mime='text/csv'
    )

# Footer
st.markdown("---")
st.markdown("""
<div style="text-align: center; color: #666;">
    <p>Sales Analytics Dashboard | Built with Streamlit & Plotly</p>
</div>
""", unsafe_allow_html=True)
'''

# Save dashboard code with UTF-8 encoding
with open('dashboard.py', 'w', encoding='utf-8') as f:
    f.write(dashboard_code)

print("Dashboard code saved as 'dashboard.py'")
print("\n" + "=" * 60)
print("TO RUN THE DASHBOARD:")
print("=" * 60)
print("1. Install Streamlit: pip install streamlit")
print("2. Run command: streamlit run dashboard.py")
print("3. Dashboard will open in your browser at http://localhost:8501")
print("=" * 60)
```

```python
# ============================================================================
# BONUS: ADDITIONAL ANALYTICS & INSIGHTS
# ============================================================================

print("\n" + "=" * 60)
print("BONUS ANALYTICS & INSIGHTS")
print("=" * 60)

# 1. Discount Effectiveness Analysis
discount_analysis = df.groupby('Discount').agg({
    'Sales': 'mean',
    'TransactionID': 'count',
    'SatisfactionScore': 'mean'
}).reset_index()
discount_analysis.columns = ['Discount_Percent', 'Avg_Sales', 'Transactions', 'Avg_Satisfaction']

print("\n--- Discount Effectiveness ---")
print(discount_analysis)

# 2. Return Analysis by Category
return_analysis = df.groupby('ProductCategory').agg({
    'Return': 'sum',
    'TransactionID': 'count',
    'Sales': 'sum'
}).reset_index()
return_analysis['Return_Rate'] = (return_analysis['Return'] / 
                                  return_analysis['TransactionID']) * 100
return_analysis = return_analysis.sort_values('Return_Rate', ascending=False)

print("\n--- Return Rate by Category ---")
print(return_analysis[['ProductCategory', 'Return_Rate']])

# 3. Payment Method Analysis
payment_analysis = df.groupby('PaymentMethod').agg({
    'Sales': 'sum',
    'SatisfactionScore': 'mean',
    'TransactionID': 'count'
}).reset_index()
payment_analysis['Avg_Order_Value'] = payment_analysis['Sales'] / payment_analysis['TransactionID']

print("\n--- Payment Method Performance ---")
print(payment_analysis)

# 4. Create correlation heatmap
plt.figure(figsize=(12, 8))
corr_cols = ['Sales', 'Discount', 'Quantity', 'Return', 'SatisfactionScore']
corr_matrix = df[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=.5)
plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n Correlation heatmap saved as 'correlation_heatmap.png'")

# 5. Key Insights Summary
print("\n" + "=" * 60)
print(" KEY BUSINESS INSIGHTS")
print("=" * 60)

# Find best performing segment
best_segment = segment_perf['Total_Sales'].idxmax()
print(f"1. Best Performing Segment: {best_segment}")

# Find highest satisfaction region
best_region = df.groupby('Region')['SatisfactionScore'].mean().idxmax()
print(f"2. Highest Satisfaction Region: {best_region}")

# Find lowest return category
lowest_return_cat = return_analysis.loc[return_analysis['Return_Rate'].idxmin(), 'ProductCategory']
print(f"3. Lowest Return Category: {lowest_return_cat}")

# Find best channel
best_channel = channel_perf['Avg_Order_Value'].idxmax()
print(f"4. Highest AOV Channel: {best_channel}")

# Discount recommendation
optimal_discount = discount_analysis.loc[discount_analysis['Avg_Sales'].idxmax(), 'Discount_Percent']
print(f"5. Optimal Discount Level: {optimal_discount}%")

print("=" * 60)
```
